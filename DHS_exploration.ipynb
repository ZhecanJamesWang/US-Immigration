```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
# plt.rcParams['figure.figsize'] = (10.0, 8.0)

def import_data():
    '''Import all data from csv files'''
    global border_patrol, investigations, apprehended, immigration
    global lpr, naturalization, prosecutions, refugees, removed, immig_reg
    border_patrol = pd.read_csv('DHS_Data/border_patrol.csv', index_col=0)
    investigations = pd.read_csv('DHS_Data/investigations.csv', index_col=0)
    apprehended = pd.read_csv('DHS_Data/apprehended.csv', index_col=0)
    immigration = pd.read_csv('DHS_Data/immigration.csv', index_col=0)
    lpr = pd.read_csv('DHS_Data/lpr.csv', index_col=0)
    naturalization = pd.read_csv('DHS_Data/naturalization.csv', index_col=0)
    prosecutions = pd.read_csv('DHS_Data/prosecutions.csv', index_col=0)
    refugees = pd.read_csv('DHS_Data/refugee.csv', index_col=0)
    removed = pd.read_csv('DHS_Data/removed.csv', index_col=0)
    immig_reg = pd.read_csv('DHS_Data/immigration_reg.csv', index_col=0)
    
import_data()
```
