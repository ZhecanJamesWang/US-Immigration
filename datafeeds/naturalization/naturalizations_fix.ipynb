```python
import pandas as pd

data86 = pd.read_excel('naturalizations_reg86-04.xls')
data04 = pd.read_excel('naturalizations_reg04-13.xls')
```

```python
data86.drop()
```

```python
data04[data04.Country=='Venezuela']
```

```python
data = pd.DataFrame(columns=data86.columns+data04.columns[1:])
data.columns
for country in data86.Country:
    begin = data86[data86.Country==country]
    end = []
    try: 
        end = data04[data04.Country==country]
    except:
        pass
    
    if end.empty:
        new_entry = begin.iloc[0].tolist() + [None,None,None,None,None,None,None,None,None]
    else:
        new_entry = begin.iloc[0].tolist() + end.iloc[0].tolist()[2:]
        
    appension =  pd.DataFrame(new_entry).transpose()
    appension.columns = data.columns
   
    data = data.append(appension)
```

```python
for country in data04.Country:
    if country not in data.Country.tolist():
        begin = [country, None,None,None,None,None,None,None,None,None,None,None,None,None,None]
        end = data04[data04.Country==country]
        
        new_entry = begin + end.iloc[0].tolist()[2:]
        
        appension =  pd.DataFrame(new_entry).transpose()
        appension.columns = data.columns
        
        data = data.append(appension)
        
trans_data = data.transpose().head()
trans_data.columns=trans_data.iloc[0]
trans_data = trans_data[trans_data.Total!='Total']

data.to_csv('../Data/naturalization_by_nat.csv')
```
